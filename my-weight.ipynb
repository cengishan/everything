# BMR = 66 + (13,7 × waga w kg) + (5 × wysokość w cm) - (6,8 × wiek w latach) x PAL

Liczba spożywanych kalorii, a waga ciała.

```python
x = 115
y= (66 + (13 * x) + (5 * 182) - (6.8 * 30))*1.5
```

```python
def compute(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = (66 + (13 * values[i]) + (5 * 182) - (6.8 * 30))*1.5 
    return output
```

```python
x = [80,85,90,95,100,105,110,115,120,125,130]
df={"weight": x,
    "bmr": compute(x)
   }

df = pd.DataFrame(df)
```

```python
df.plot(x="weight",y="bmr", title="liczba spożywanych kalori, a waga ciała")
```

Jak zmienia się procent tkanki tłuszczowej (BF) oraz współczynnik BMI od mojej wagi ciała. BMI <30 lub BF < 30% oznaczają brak otyłości. Zaznaczono je kolorem czerwonym. 

```python
x = 115
y= 70 / (70 + 45)
```

```python
def compute(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1-(70 / values[i])
    return output
```

```python
x = [80,85,90,95,100,105,110,115,120,125,130]
df={"weight": x,
    "bf": compute(x)
   }

df = pd.DataFrame(df)
```

```python
plot=df.plot(x="weight",y="bf", title="procent tkanki tłuszczowej")
plot.axhline(y=0.30, xmin=-1, xmax=1, color='r', linestyle='--', lw=2)
```

```python
def compute(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = values[i]/(1.82**2)
    return output
```

```python
x = [80,85,90,95,100,105,110,115,120,125,130]
df={"weight": x,
    "bmi": compute(x)
   }

df = pd.DataFrame(df)
```
