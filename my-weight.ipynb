# BMR = 66 + (13,7 √ó waga w kg) + (5 √ó wysoko≈õƒá w cm) - (6,8 √ó wiek w latach) x PAL

Liczba spo≈ºywanych kalorii, a waga cia≈Ça.

```python
x = 115
y= (66 + (13 * x) + (5 * 182) - (6.8 * 30))*1.5
```

```python
def compute(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = (66 + (13 * values[i]) + (5 * 182) - (6.8 * 30))*1.5 
    return output
```

```python
x = [80,85,90,95,100,105,110,115,120,125,130]
df={"weight": x,
    "bmr": compute(x)
   }

df = pd.DataFrame(df)
```

```python
df.plot(x="weight",y="bmr", title="liczba spo≈ºywanych kalori, a waga cia≈Ça")
```

Jak zmienia siƒô procent tkanki t≈Çuszczowej (BF) oraz wsp√≥≈Çczynnik BMI od mojej wagi cia≈Ça. BMI <30 lub BF < 30% oznaczajƒÖ brak oty≈Ço≈õci. Zaznaczono je kolorem czerwonym. 

```python
x = 115
y= 70 / (70 + 45)
```

```python
def compute(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1-(70 / values[i])
    return output
```

```python
x = [80,85,90,95,100,105,110,115,120,125,130]
df={"weight": x,
    "bf": compute(x)
   }

df = pd.DataFrame(df)
```

```python
plot=df.plot(x="weight",y="bf", title="procent tkanki t≈Çuszczowej")
plot.axhline(y=0.30, xmin=-1, xmax=1, color='r', linestyle='--', lw=2)
```
